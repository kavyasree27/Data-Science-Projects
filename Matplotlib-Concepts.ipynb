## Basic Example 

We'll create two numpy arrays and try to plot

```python
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
```

## Basic matplotlib commands

```python
plt.plot(x,y,'r');
plt.xlabel('This is X-Axis')
plt.ylabel('This is Y-Axis')
plt.title('This is plot title')
plt.show();
```

## Create Multiplots on same canvas


**plt.subplot(arg1,arg2,arg3)**

* arg1 : n-rows
* arg2 : n-cols
* arg3 : plot number

```python
plt.subplot(1,2,1) 
plt.plot(x,y,'r--')
plt.subplot(1,2,2) 
plt.plot(y,x,'g+-');
#plt.show()
```

```python
x=range(10)
y=range(10)

fig,ax = plt.subplots(nrows=2,ncols=2)

for row in ax:
    for col in row:
        col.plot(x,y)

plt.show()
```

## Matplotlib object oriented approach

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.7,0.7]) # left, bottom, width, height (range is between 0 & 1)

# let's plot on the axes
axes.plot(x,y,'b')
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('Title')
```

```python
# Create a blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9])  # main axis
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])  # inset axis

# let's plot on the axes
axes1.plot(x,y,'b')
axes1.set_xlabel('X-axis')
axes1.set_ylabel('Y-axis')
axes1.set_title('Axes 1 Title')

axes2.plot(y,x,'r-')
axes2.set_xlabel('X-axis')
axes2.set_ylabel('Y-axis')
axes2.set_title('Axes 2 Title')
```
